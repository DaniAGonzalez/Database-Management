**Importing libraries**

```python
import sqlite3
import pandas as pd
```

**Retrieving the information from different types of files extension **



```python
def ReadDB():
  conn = sqlite3.connect('/content/sales_data.db')
  df1 = pd.read_sql(f'SELECT * FROM sales', conn)
  return df1
```

```python
def ReadTex():
  df2 = pd.read_csv('/content/sales_data.txt', sep='\t')
  return df2
```

```python
def ReadCSV():
  df3 = pd.read_csv('/content/sales_data.csv')
  return df3
```

```python
df1=(ReadDB())
```

```python
df2 =(ReadTex())
```

```python
df3=(ReadCSV())
```

**Concatenation of the three df**


```python
final_df = pd.concat([df1, df2, df3], ignore_index=True)
final_df.to_csv('final_data.csv', index=False)
final_df
```

**Checking missing values**

```python

# Checking if missing values
def Miss_Data(final_df):
  final_df['Sales'].fillna(final_df ['Sales'].mean(), inplace=True) #Checking missing values for the Y variables
  final_df['Profit'].fillna(final_df['Profit'].mean(), inplace=True) #Checking missing values for the Y variables
  final_df.dropna(subset=['Branch'], inplace=True) #Checking missing values for the X variables
  return final_df

CleanData = Miss_Data(final_df)
```

```python
# Function to check doubled data
def DoubleData(CleanData):
  CleanDF = CleanData.drop_duplicates()
  return CleanDF

Cleanall=DoubleData(CleanData)
```

```python
# Checking outliers
```

```python
# Normalization
def normalizedata(Cleanall):
  Cleanall['Sales'] = (Cleanall['Sales'] - Cleanall['Sales'].mean()) / (Cleanall['Sales'].std())
  Cleanall['Profit'] = (Cleanall['Profit'] - Cleanall['Profit'].mean()) / (Cleanall['Profit'].std())
  return Cleanall

# Call normalizedata and assign the result to Final_clean
Final_clean = normalizedata(Cleanall)
print(Final_clean.head(4))

# We can add a normalization also using the min and max values
```

```python
Analysis_Result = Final_clean.groupby('Sales').sum().reset_index()
Analysis_Result.head(50)
```

```python
# Retrieving the sales based on the time
Final_clean['Data']= pd.to_datetime(Final_clean['Date'])
Final_clean['YearMonth'] = Final_clean['Data'].dt.to_period('M')
Final_monthly_sales_trend = Final_clean.groupby('YearMonth')['Sales'].sum().reset_index()
print(Final_monthly_sales_trend)
```

```python
# Sales based on the branch
Final_branch_sales = Final_clean.groupby('Branch')['Sales'].sum().reset_index()
print(Final_branch_sales)
```

```python
# 1) What are the total sales for each branch in January 2024?


TotalSales = Final_clean[Final_clean['YearMonth'] == '2024-01'].groupby('Branch')['Sales'].sum().reset_index()
print(TotalSales)
```

```python
# 2) Sales for each month
MonthlySales = Final_clean.groupby('YearMonth')['Sales'].sum().reset_index()
print(MonthlySales)
```

```python
# 3)Which branch had the highest sales in March 2024?
HighestSales = Final_clean[Final_clean['YearMonth'] == '2024-03'].groupby('Branch')['Sales'].max().reset_index()  # Filter for March 2024 and use groupby
# Get branch with max sales using idxmax
highest_sales_branch = HighestSales.loc[HighestSales['Sales'].idxmax(), 'Branch']
print(highest_sales_branch)
```

```python
# 4) What is the average monthly profit for each branch in the dataset?
# Calculate the average monthly profit for each branch
AverageProfit = Final_clean.groupby(['Branch', 'YearMonth'])['Profit'].mean().reset_index()
print(AverageProfit)
```

```python
# 5) Identify the month with the highest total sales across all branches.
MaxSales = Final_clean.groupby('YearMonth')['Sales'].sum().idxmax()
print(MaxSales)
```

```python
# 6) Determine the month in which Branch C had the highest sales.
# Filter the DataFrame for Branch C first
BranchC_data = Final_clean[Final_clean['Branch'] == 'Branch C']
# Then, find the month with the highest sales
branchC_highest = BranchC_data.groupby('YearMonth')['Sales'].sum().idxmax()
print(branchC_highest)
```

```python

#8) Which branch had the most negative profit in the dataset, and in which month did this occur?
# Group by Branch and YearMonth and calculate the sum of profits
branch_negative_profit = Final_clean.groupby(['Branch', 'YearMonth'])['Profit'].sum()

# Find the branch and month with the most negative profit
most_negative_profit = branch_negative_profit.idxmin()
lowest_profit_value = branch_negative_profit.min()

print(lowest_profit_value)
print(most_negative_profit)
```

```python
# 9) Calculate the cumulative sales trend for Branch A over the period covered in the dataset.
branch_a_data = Final_clean[Final_clean['Branch'] == 'Branch A']
branch_a_data = branch_a_data.sort_values(by='Date')
branch_a_data['Cumulative_Sales'] = branch_a_data['Sales'].cumsum()


print(branch_a_data[['Date', 'Sales', 'Cumulative_Sales']])
```

```python
# 10) Compare the sales trends between Branch A and Branch B for the first quarter of 2024 (January to March).
quarter_filter = Final_clean[(Final_clean['YearMonth'] >= '2024-01') & (Final_clean['YearMonth'] <= '2024-03')]
branch_a_sales = quarter_filter[quarter_filter['Branch'] == 'Branch A']['Sales'].sum()
branch_b_sales = quarter_filter[quarter_filter['Branch'] == 'Branch B']['Sales'].sum()

print(f' Branch A sales: {branch_a_sales}')

print(f' Branch B sales: {branch_b_sales}')

```
